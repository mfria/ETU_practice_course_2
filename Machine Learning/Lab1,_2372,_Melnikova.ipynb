# Набор данных о студенческом образе жизни

**Описание датасета**:
Этот набор данных дает подробное представление о моделях образа жизни учащихся и их корреляции с академической успеваемостью, представленной средним баллом успеваемости.

**Предметная область**: Образование

**Источник данных**: https://www.kaggle.com/datasets/steve1215rogg/student-lifestyle-dataset

**Характер данных**: реальные

**Атрибуты датасета**: Идентификатор студента, количество учебных часов в день,время дополнительной активности в день, время сна в день, время социальной активности в день, часы физической активности в день, учебный балл, уровень стресса. **Тип представленных данных** - числовой.

**Задача анализа**:
Выявить зависимость успеваемости - среднего балла студентов от количества часов учёба в день

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('student_lifestyle_dataset.csv', delimiter=',')

# Вычисление среднего значения
mean_Sleep_Hours = df['Study_Hours_Per_Day'].mean()
std_Sleep_Hours = df['Study_Hours_Per_Day'].std()
print("Атрибут 'Study_Hours_Per_Day' - количество часов учебы в сутки")
print("Среднее значение атрибута 'Study_Hours_Per_Day': ", mean_Sleep_Hours)
print("Среднее квадратичное отклонение атрибута 'Study_Hours_Per_Day': ", std_Sleep_Hours)

# Вычисление среднеего квадратичного отклонения
mean_GPA = df['GPA'].mean()
std_GPA = df['GPA'].std()
print("\nАтрибут 'GPA' - средний балл")
print("Среднее значение атрибута 'GPA': ", mean_GPA)
print("Среднее квадратичное отклонение атрибута 'GPA': ", std_GPA)
```

```python
# Построение гистограмм
plt.hist(df['Study_Hours_Per_Day'], bins = 50)
plt.xlabel('Количество часов учебы в сутки')
plt.ylabel('Частота')
plt.title("Распределение атрибута 'Study_Hours_Per_Day'")
plt.show()

plt.hist(df['GPA'], bins = 50)
plt.xlabel('Средний балл')
plt.ylabel('Частота')
plt.title("Распределение атрибута 'GPA'")
plt.show()
```

Из гистограмм видим, что выбросов данных у исследуемых атрибутов нет.

```python
# Проверка на наличие пропущенных значений
missing_values = df.isnull().sum()
print("Количество пропущенных значений:")
print(missing_values)
```

Пропущенных значений у нет.

```python
# Корреляция между параметрами
correlation = df[['Study_Hours_Per_Day', 'GPA']].corr()

print("Корреляция между количеством количество часов учебы в сутки 'Study_Hours_Per_Day' и средним баллом 'GPA' :")
print(correlation)
```

Значение корреляции 0.734468 положительное, по модулю блишзкое к 1 означает, что зависимость прямая и достаточно сильная: чем больше учебных часов в сутки, тем выше средний балл.

```python
# Построение графика рассеивания
sns.scatterplot(x='Study_Hours_Per_Day', y='GPA', data=df)
plt.title('График рассеивания')
plt.show()
```

График рассеивания позволяет визуально оценить взаимосвязь между значениями этих атрибутов. Из графика видно, что взаимосвязь ярко выражена, точки на графике имеют тенденцию к расположению на прямой вида некоторой линейной функции.
